# choosing the right estimator/algorithm for yor problem

Some things to note:
    * Sklearn refers to machine learning models as estimators

```python
# Get calforinaia dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing
```

```python
housing_df = pd.DataFrame(housing["data"],columns=housing["feature_names"])
housing_df
```

```python
housing_df["MedHouseVal"] = housing["target"]
housing_df
```

```python
# import Algorithm
# Setup randomseed(42)

#  Create the data
X = housing_df.drop("MedHouseVal",axis=1)
Y = housing_df["MedHouseVal"]

# Split into train and test data
X_Train,X_Test,Y_Train,Y_Test = train_test_split(X,Y,test_size=0.2)

# Instantiate and fit the model (on training set)
from sklearn import linear_model
model = linear_model.Ridge()
model.fit(X_Train,Y_Train)

#  check model score
model.score(X_Test,Y_Test)
```

 What if 'Ridge' didn't work or score didn't fit our needs?
 Well , we could always try a different model...
 
 how about we try an ensemble modle(an ensemble is combination of smaller models to try and make better prediction than just a single algorithm


```python
#  import Randomforestregressor model class from the ensemble module
from sklearn.ensemble import RandomForestRegressor
np.random.seed(42)

model = RandomForestRegressor(n_estimators=100)
model.fit(X_Train,Y_Train)
model.score(X_Test,Y_Test)
```

```python
model.predict(X_Test)
```
