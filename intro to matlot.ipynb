# MatPlotLib Example Workflow

```python
# 0. import matplotlib

# 1. Prepare Data
x= [1,2,3,4]
y = [11,22,33,44]

# 2. create plot
fig , ax = plt.subplots(figsize=(10,10))

# 3. Plot Data
ax.plot(x,y)

# 4. setup plot
ax.set(title = "First Plot",
      xlabel="x axis"
      ,ylabel="y axis")
# 5. save figure
fig.savefig("firstplot.png")
```

# Making figures with Numpy arrays
We want:
* Line Plot
* Scatter Plot
* Bar Plot
* Histogram Plot

```python
x = np.linspace(0,10,50)
y = np.linspace(10,0,50)
plt.plot(x,'gs',y,"bs")
```

```python
fig, ax= plt.subplots()
ax.plot(x,x**5);
ax.set(title= "POWER OF 5",xlabel="value of data",ylabel="their power")
```

# make a plot from dictionary

```python
nut_butter_prices = {"Almond Butter":10,
                     "Peanut Butter":8,
                     "Cashew Butter":12
}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(xlabel = "different types",ylabel = "their prices in $")
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
```

```python
x = np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x)
```

# Two Options for subplots

```python
fig ,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)
ax1.plot(x,x/2)
ax2.hist(x)
ax3.barh(x,x**2)
ax4.scatter(x,np.tan(x))
```
